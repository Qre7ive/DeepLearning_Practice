```python
# 创建一个人工数据集

import os

os.makedirs(os.path.join('..','data'),exist_ok=True)
# 创建一个名为 data 的文件夹（目录），路径在当前工作目录的上一级目录（..）
# 在当前 Colab/项目目录的上一层创建一个 /data 文件夹，例如

data_file = os.path.join('..','data','house_tiny.csv')
# 定义了要创建的CSV文件路径：上级目录的data文件夹下的house_tiny.csv

# 写入数据
with open(data_file,'w') as f:# 以写入模式（'w'）打开文件
  # 第一行写入CSV表头：NumRooms（房间数量）、Alley（小巷or路的类型）、Price（价格）
  f.write('NumRooms,Alley,Price\n')
  # 后续几行写入具体数据，其中NA表示缺失值
  f.write('NA,Pave,127500\n')
  f.write('2,NA,10600\n')
  f.write('4,NA,178100\n')
  f.write('NA,NA,140000\n')
```

```python
import pandas as pd
# 读取数据集
data = pd.read_csv(data_file)
print(data)
data
# 不print是用html文件显示数据
```

```python
# 为了处理缺失的数据，典型的方法包括插值和删除，这里将考虑插值
inputs = data.iloc[:,0]
# 令inputs为第0
inputs = inputs.fillna(inputs.mean())
# fillna()将NaN填充为这一列的均值
print(inputs)
```

```python
inputs, outputs = data.iloc[:,0:2], data.iloc[:,2]
inputs1 = data.iloc[:,0]
data.iloc[:,0] = data.iloc[:,0].fillna(inputs1.mean())# 用均值填充NumRooms列的NAN
print(inputs)
```

```python
# 使用get_dummies()将“类别型变量”转换为 哑变量（dummy variables）
# 对于inputs的类别值或离散值，我们将NaN视为一个类别
inputs = pd.get_dummies(inputs,dummy_na=True,dtype=int)# 若为False,则没有Alley_nan列
# dummy_na=True 指的是 是否为缺失值（NaN）单独创建一列
print(inputs)
```
