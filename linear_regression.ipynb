```python
A = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
# 计算总和或均值的时候保持轴数不变
sum_A = A.sum(axis=1, keepdims=True)
# keepdims=True让结果仍然保持原来的维度数量
sum_A, sum_A.shape
```

```python
# 因为形状相同所以可以运算
A / sum_A
```

```python
# 某个轴计算A元素的累积总和
A.cumsum(axis=0)
```
